```python
    _, threshold = cv.threshold(image, 180, 255, cv.THRESH_BINARY)
    kernel = np.ones((5, 5), np.uint8)
    closing = cv.morphologyEx(threshold, cv.MORPH_CLOSE, kernel)

    mask = cv.bitwise_not(closing)
    result = cv.bitwise_and(image, image, mask=mask)
    plt.imshow(result)
```

```python
if image.shape == result.shape:
    subtracted_image = cv.subtract(imagem, result)
```

```python
plt.imshow(subtracted_image)
```

```python
image = (255-subtracted_image)
plt.imshow(image, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
noise_image=image
erosion = cv.erode(noise_image, kernel, iterations=1)
plt.imshow(erosion,cmap='gray')
```
